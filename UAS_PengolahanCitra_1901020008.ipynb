RESEARCH : EROSION DILATION

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('bnw.jpg', 0)
```

```python
matrixFilter = np.ones((5,5), np.uint8)
```

```python
erosion = cv2.erode(img, matrixFilter, iterations=1)
dilation = cv2.dilate(img, matrixFilter, iterations=1)

titles = ['original', 'erosion', 'dilation']
images = [img, erosion,  dilation]

for i in range(3):
  plt.subplot(2,2, i + 1)
  plt.title(titles[i])
  plt.imshow(images[i])
```

```python
histogram = cv2.calcHist([img], [0], None, [256], [0,256])
plt.plot(histogram)
```

**RESEARCH : IMAGE FILTER**

```python
from PIL import Image
from urllib.request import urlopen
from numpy import asarray
from PIL import ImageFilter
```

```python
url = 'https://www.delftstack.com/img/Python/image%20example.jpg?ezimgfmt=rs:640x426/rscb5/ng:webp/ngcb5'
img = Image.open(urlopen(url))
numpyData = asarray(img)
print(numpyData.shape)
```

```python
img
```

```python
grayscale = img.convert('L')
```

```python
grayscale
```

```python
tresh = 200 
fn = lambda x : 255 if x > tresh else 0
r = grayscale.point(fn, mode ='1')
```

```python
r
```

```python
for i in range(0, img.size[0]-1):

    for j in range(0, img.size[1]-1):

        # Get pixel value at (x,y) position of the image

        pixelColorVals = img.getpixel((i,j));

        #(192,159,232)

        # Invert color

        redPixel    = 255 - pixelColorVals[0]; # Negate red pixel

        greenPixel  = 255 - pixelColorVals[1]; # Negate green pixel

        bluePixel   = 255 - pixelColorVals[2]; # Negate blue pixel

                   

        # Modify the image with the inverted pixel values

        img.putpixel((i,j),(redPixel, greenPixel, bluePixel));
```

**RESEARCH : IMAGE COLOR TO BINARY**

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('widiastra.jpg')
```

```python
from IPython.core.display import Image
val, threshBinary = cv2.threshold(image, 120,120, cv2.THRESH_BINARY)
val, threshBinaryInverted = cv2.threshold(image, 120,120, cv2.THRESH_BINARY_INV)
val, threshTrunc = cv2.threshold(image, 120,120, cv2.THRESH_TRUNC)
val, threshToZero= cv2.threshold(image, 120,120, cv2.THRESH_TOZERO)
val, threshToZeroInv= cv2.threshold(image, 120,120, cv2.THRESH_TOZERO_INV)
```

```python
titles = ["Binary", "Binary Inverted", "Truncate", "TO Zero", "To Zero Inverted", "original"]
images = [threshBinary, threshBinaryInverted, threshTrunc, threshToZero, threshToZeroInv, image]

for i in range(6):
  plt.subplot(3,2, i + 1)
  plt.title(titles[i])
  plt.imshow(images[i])
```

**RESEARCH RESIZE IMAGE**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
circles = plt.imread("widiastra.jpg")
plt.imshow(circles)
plt.show()
```

```python
cat = cv2.imread("widiastra.jpg")[:, :, ::-1]
center = (1024, 1024)
M = cv2.getRotationMatrix2D(center, 0, 0.3)
resized_cat = cv2.warpAffine(cat, M, (cat.shape[1], cat.shape[0]))
plt.imshow(cat)
plt.figure()
plt.imshow(resized_cat)
```

**RESEARCH : BACKGROUND SUBSTRACTION**

```python
import cv2 
from google.colab.patches import cv2_imshow
```

```python
capture = cv2.VideoCapture('cctv.mp4')
fgbg = cv2.createBackgroundSubtractorMOG2()

while(1):
  ret, frame = capture.read()
  fgmask = fgbg.apply(frame)

  cv2_imshow( fgmask)
  cv2_imshow( frame)

  k = cv2.waitKey(30) & 0xff
  
  
capture.release()
```

**RESEARCH : IMAGE COLLORING**

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('widiastra.jpg');
cv2_imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
blur = cv2.GaussianBlur(img_gray, (3,3), 0)
```

```python
cv2_imshow(blur)
```

```python
sobel = cv2.Sobel(blur, cv2.CV_64F, 1, 1, 5)
```

```python
cv2_imshow(sobel)
```

```python
canny = cv2.Canny(blur, 200, 200)
```

```python
cv2_imshow(canny)
```
